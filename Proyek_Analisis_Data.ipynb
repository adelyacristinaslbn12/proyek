## Visualization & Explanatory Analysis

### Pertanyaan 1:
Berapa persentase rata-rata jumlah pengguna sepeda berdasarkan cuaca?

```python
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
data_by_month = day_df.groupby(day_df['dteday'].dt.to_period('M')).agg({'cnt':'sum'}).reset_index()
```

```python
result_mean_by_weathersit = day_df.groupby(by="weathersit")["cnt"].mean().reset_index()
result_mean_by_weathersit
```

```python
labels = ("clear","mist"," light rain/snow")
plt.pie(
    x=result_mean_by_weathersit["cnt"],
    labels=labels,
    autopct='%1.1f%%'
)
plt.show()
```
